```python
df_drop = ['ADDRESSLINE1', 'ADDRESSLINE2', 'STATUS','POSTALCODE',
           'CITY','TERRITORY', 'PHONE', 'STATE', 'CONTACTFIRSTNAME',
           'CONTACTLASTNAME', 'CUSTOMERNAME', 'ORDERNUMBER']
df = df.drop(df_drop, axis=1)
```

```python
country = pd.get_dummies(df['COUNTRY'])
productline = pd.get_dummies(df['PRODUCTLINE'])
Dealsize = pd.get_dummies(df['DEALSIZE'])
```

```python
df = pd.concat([df,country,productline,Dealsize], axis = 1)
```

```python
df.head()
```

```python
df_drop = ['COUNTRY','PRODUCTLINE','DEALSIZE']
df = df.drop(df_drop, axis=1)
```

```python
df.dtypes
```

```python
df['PRODUCTCODE'] = pd.Categorical(df['PRODUCTCODE']).codes
```

```python
df.dtypes
```

```python
df.drop('ORDERDATE', axis=1, inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.cluster import KMeans
# Assuming df is your DataFrame with mixed data types
numeric_columns = df.select_dtypes(include=[np.number]) # Select only numeric columns
WCSS = [] # Within Cluster Sum of Squares from the centroid
distortions = []
K = range(1, 10)

for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(numeric_columns) # Fit the model on numeric data
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16, 8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeanModel = KMeans(n_clusters=3)
y_kmeans = kmeanModel.fit_predict

print(y_kmeans)

plt.figure(figsize = (30,26))
sns.heatmap(df.corr(),annot = True)
```
